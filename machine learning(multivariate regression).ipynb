```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def multivariate_regression(x,y,alpha=0.001,num_iter=200):
    n = len(y)
    a = b = c = 0
    theta = np.array([a,b,c])
    
    
    for i in range(num_iter):
        hypothesis_func = np.dot(x,theta) # dim 5,3 * 3,1 = 1,5
        cost = (1/n)*sum((y-hypothesis_func)**2)
        gradient_a = (-2/n)*np.sum((y-hypothesis_func)*x[:,0])
        gradient_b = (-2/n)*np.sum((y-hypothesis_func)*x[:,1])
        gradient_c = (-2/n)*np.sum((y-hypothesis_func)*x[:,2])
        temp_a = a - alpha*gradient_a
        temp_b = b - alpha*gradient_b
        temp_c = c - alpha*gradient_c
        a = temp_a
        b = temp_b
        c = temp_c
        theta = np.array([a,b,c])
        print("a {}, b {}, c {}, cost {} ,  num_iter {}".format(a,b,c,cost,i))
        
    y_predicted = np.dot(x,theta)
    
    plt.scatter(x[:,1],y)
    plt.scatter(x[:,2],y)
    plt.plot(x[:,1],y_predicted,'-g')
    plt.plot(x[:,2],y_predicted,'-g')
    
        
    return a,b,c,cost

def predict(x,theta):
    y_predicted = np.dot(x,theta)
    return y_predicted
```

```python
x = np.array( [ [1,2,8],
                [1,3,11],
                 [1,5,13],
                [1,6,15],
               [1,8,16]  ]  )

y = np.array([30,33,36,39,41])

multivariate_regression(x,y)
```
