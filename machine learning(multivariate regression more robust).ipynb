```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost_function(x,y,theta):
    n = len(y)
    H = np.dot(x,theta)
    cost = (1/n)*np.sum((y-H)**2)
    return cost

def gradient_descent(x,y,theta,alpha,num_iter):
    n = len(y)
    cost_list =[]
    for i in range(num_iter):
        H = np.dot(x,theta)
        print(H.shape)
        print(y.shape)
        error = (y-H)
        gradient = (-2/n)*(error.T.dot(x))
        temp_0 = theta - alpha*gradient
        theta = temp_0
        cost = cost_function(x,y,theta)
        cost_list.append(cost)
        print("theta {} , cost {}, num_iter {}".format(theta,cost,i))
        #print(cost_list)
        
    y_predicted = np.dot(x,theta)
    print(y_predicted)
    
    y_predicted = np.dot(x,theta)
    plt.scatter(x[:,1],y)
    plt.scatter(x[:,2],y)
    plt.plot(x[:,1],y_predicted,'-g')
    plt.plot(x[:,2],y_predicted,'-g')
    
    return theta
    
```

```python
x = np.array( [ [1,2,8],
                [1,3,11],
                 [1,5,13],
                [1,6,15],
               [1,8,16]  ]  )

y = np.array([30,33,36,39,41])
theta = np.array([0,0,0])
gradient_descent(x,y,theta,alpha=0.001,num_iter=100)


```
