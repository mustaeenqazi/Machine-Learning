```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

 generating data to use for simple linear regression.

```python
x = np.array([2,4,6,8,10,12,13,15,16,20,22,25,26])
y = np.array([11,14,17,20,24,28,30,33,34,38,40,41,47])
plt.plot(x,y,'.')
plt.xlabel("xvals")
plt.ylabel("yvals")
print(np.size(x))
print(np.size(y))
```

```python
def linear_regression(x,y,alpha = 0.01,num_iter = 100):

    n = len(x)        #number of values 'n' = np.size(x)
    a = b = 0       # first give intial values to a and b
#hypothesis function = h_x = b+a*x
#cost function= j_a_b = (1/2*n)*np.sum((h_x-y)^2) 
#we need to take partial derivative of this cost function w.r.t a and b
#j_b = 1/n*np.sum((h_x-y)^2) # partial derivative w.r.t 'b'
#j_a = 1/n*np.sum(((h_x-y)^2)*x) # partial derivative w.r.t 'a'# alpha is the gradient descent coefficient(learning rate)
    #alpha = 0.01
    #num_iter = 100 #number of iterations
    cost_list = []

    for i in range(num_iter):
        hypothesis_func = b + a*x
        cost = (1/n)*sum([data**2 for data in (y-hypothesis_func)])
        cost_list.append(cost)
        gradient_b = (-2/n)*np.sum((y-hypothesis_func))
        gradient_a = (-2/n)*np.sum((y-hypothesis_func)*x)
        temp_b = b - alpha*gradient_b
        temp_a = a - alpha*gradient_a
        b = temp_b
        a = temp_a
        print("b {}, a {},cost {}, num_iter {}".format(b,a,cost,i)) # this function is used to print values at each iteration
        
    y_predicted = b + a*x 
    
                                 #plt.plot(list(range(num_iter)), cost_list, '-r') 
    plt.scatter(x,y)
    plt.plot(x,y_predicted,'-g')
    plt.show()
    print(np.array(cost_list))
    print(hypothesis_func)
    
    return b,a, cost, cost_list  

def predict(x,a,b):
   
    y = b+a*x
    return y
  

```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

linear_regression(x,y)
```

```python
x = np.array([2,4,6,8,10,12,13,15,16,20,22,25,26])
y = np.array([11,14,17,20,24,28,30,33,34,38,40,41,47])
linear_regression(x,y,alpha=0.001,num_iter=100)
predict(5,1.9,0.5)
```
